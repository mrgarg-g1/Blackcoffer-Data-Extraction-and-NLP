```python
import os

def load_stopwords(file_path, encoding='utf-8'):
    with open(file_path, 'r', encoding=encoding, errors='ignore') as file:
        stopwords = [line.strip() for line in file]
    return stopwords
# Assuming your script and stopwords files are in the same directory
script_directory = os.getcwd()

# Load stopwords from each file
stopwords_auditor = load_stopwords(os.path.join(script_directory, 'StopWords_Auditor.txt'))
stopwords_currencies = load_stopwords(os.path.join(script_directory, 'StopWords_Currencies.txt'))
stopwords_dates_numbers = load_stopwords(os.path.join(script_directory, 'StopWords_DatesandNumbers.txt'))
stopwords_generic = load_stopwords(os.path.join(script_directory, 'StopWords_Generic.txt'))
stopwords_generic_long = load_stopwords(os.path.join(script_directory, 'StopWords_GenericLong.txt'))
stopwords_geographic = load_stopwords(os.path.join(script_directory, 'StopWords_Geographic.txt'))
stopwords_names = load_stopwords(os.path.join(script_directory, 'StopWords_Names.txt'))

# Combine all stopwords into a single list
all_stopwords = (
    stopwords_auditor + stopwords_currencies + stopwords_dates_numbers +
    stopwords_generic + stopwords_generic_long + stopwords_geographic +
    stopwords_names
)
```

```python
# clean and tokenize
def clean_and_tokenize(text):
    stop_words = all_stopwords
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    return tokens
```

```python
with open('positive-words.txt', 'r', encoding='latin1') as file:
    positive_words = [line.strip() for line in file]
with open('negative-words.txt', 'r', encoding='latin1') as file:
    negative_words = [line.strip() for line in file]


def text_analysis(text,positive_words, negative_words):
    tokens = clean_and_tokenize(text)
    positive_score = sum(1 for token in tokens if token in positive_words)
    negative_score = sum(1 for token in tokens if token in negative_words) * -1  # Multiply by -1 for negative scoring
    polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 1e-6)
    subjectivity_score = (positive_score + negative_score) / (len(tokens) + 1e-6)
    avg_sentence_length = len(tokens) / len(sent_tokenize(text))
    percentage_of_complex_words = sum(len(word) > 2 for word in tokens) / len(tokens)
    fog_index = 0.4 * (avg_sentence_length + percentage_of_complex_words)
    word_count = len(tokens)
    complex_word_count = sum(len(word) > 2 for word in tokens)


    def syllable_count_per_word(word):
        if word.endswith("es") or word.endswith("ed"):
            return max(1, count_vowels(word[:-2]))  # Exclude "es" or "ed" before counting vowels
        else:
          return max(1, count_vowels(word))

    def count_vowels(word):
        return sum(1 for char in word if char.lower() in 'aeiou')

    for i in tokens:
      syllable_count_per_word(i)


    personal_pronouns = sum(1 for word in tokens if word.lower() in ['i', 'we', 'my', 'ours', 'us'] and word.lower() not in ["us"] )
    avg_word_length = sum(len(word) for word in tokens) / word_count
    avg_words_per_sentence = word_count / len(sent_tokenize(text))


    results = {
        'POSITIVE SCORE': positive_score,
        'NEGATIVE SCORE': negative_score,
        'POLARITY SCORE': polarity_score,
        'SUBJECTIVITY SCORE': subjectivity_score,
        'AVG SENTENCE LENGTH': avg_sentence_length,
        'PERCENTAGE OF COMPLEX WORDS': percentage_of_complex_words,
        'FOG INDEX': fog_index,
        'WORD COUNT': word_count,
        'COMPLEX WORD COUNT': complex_word_count,
        'SYLLABLE PER WORD': syllable_count_per_word,
        'PERSONAL PRONOUNS': personal_pronouns,
        'AVG WORD LENGTH': avg_word_length,
        'AVG NUMBER OF WORDS PER SENTENCE': avg_words_per_sentence
    }

    return results

for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    title = soup.find('title').get_text()
    article_text = ' '.join([p.get_text() for p in soup.find_all('p')])

    with open(f'{url_id}.txt', 'w', encoding='utf-8') as file:
        file.write(f'Title: {title}\n\n')
        file.write(article_text)


    analysis_results = text_analysis(article_text, positive_words, negative_words)


    for key, value in analysis_results.items():
        df.at[index, key] = value

df.to_excel("Blackcoffer Data Extraction and NLP.xlsx", index=False)
```
